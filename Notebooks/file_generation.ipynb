```python
from dotenv import load_dotenv
import os
from openai import OpenAI
from IPython.display import Markdown, display
from pathlib import Path
```

```python
load_dotenv(override=True)
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')

if openrouter_api_key:
    print(f"OpenRouter API Key exists and begins {openrouter_api_key[:3]}")
else:
    print("OpenRouter API Key not set (and this is optional)")

openrouter_url = "https://openrouter.ai/api/v1"

openrouter = OpenAI(base_url=openrouter_url, api_key=openrouter_api_key)
```

```python
MODEL = 'google/gemini-2.5-flash-lite'
file_dir = r'C:\Users\rauna\projects\llm_engineering\My Projects\Drug Chatbot\Synthetic_files'
topics = [
    "Physical appearance and distinctive features",
    "Personality traits and temperament",
    "Core values and beliefs",
    "Professional background and career path",
    "Educational history and academic achievements",
    "Hobbies and recreational interests",
    "Family background and relationships",
    "Cultural heritage and traditions",
    "Communication style and social skills",
    "Strengths and talents",
    "Weaknesses and areas for growth",
    "Life goals and aspirations",
    "Daily routines and habits",
    "Emotional intelligence and coping mechanisms",
    "Leadership style and influence on others",
    "Creative expressions and artistic pursuits",
    "Political views and social causes",
    "Health and fitness practices",
    "Travel experiences and geographic influences",
    "Key life experiences and defining moments"
]
```

```python
def generate_files_with_llm(num_files=1, output_dir=file_dir):
    print(f"Generating {num_files} files\n")

    for i in range(num_files):
        topic = topics[i % len(topics)]
        filename = f"file_{i+1:02d}.txt"
        filepath = os.path.join(output_dir, filename)
        try:
            # Call OpenAI API
            response = openrouter.chat.completions.create(
                model=MODEL,  # You can also use "gpt-4o-mini" for faster/cheaper generation
                messages=[
                    {
                        "role": "system",
                        "content": "you are an expert in creating systhetic data for a person named Paracetamol which is a male"
                    },
                    {
                        "role": "user",
                        "content": f"Write {topic}. Make it engaging and informative."
                    }
                ],
                max_tokens=1000,
                temperature=0.7
            )
            content = response.choices[0].message.content
            with open(filepath, 'w', encoding='utf-8') as f:
                f.write(f"Topic: {topic}\n\n")
                f.write(content)
            
            print(f"Generated: {filename}")
            
        except Exception as e:
            print(f"✗ Error generating {filename}: {str(e)}")
    
    print(f"\n✓ All files saved to '{output_dir}/' directory")
```
